```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from utils import *

imputer = SimpleImputer(strategy='mean')

attr_numericos = ["Culmen Length (mm)", "Culmen Depth (mm)"
                  , "Flipper Length (mm)", "Body Mass (g)", "Delta 15 N (o/oo)", "Delta 13 C (o/oo)"]

coluna_alvo = ["Species"]
X = df[attr_numericos]
# Lidar com dados ausentes
X = MySimpleImputer(strategy="mean").fit_transform(X)
X = MyScaler().fit_transform(X)
y = df[coluna_alvo]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from knn import KNN

knn = KNN(3)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Acurácia: ", accuracy_score(y_test,y_pred))
print("Precisão: ", precision_score(y_test,y_pred, average=None))
print("Recall:   ", recall_score(y_test,y_pred, average=None))
print("F1 Score: ", f1_score(y_test,y_pred, average=None))
```
